```python
import os
import win32com.client as win32

def convert_xls_to_xlsm(directory):
    excel = win32.Dispatch('Excel.Application')
    excel.DisplayAlerts = False  # Disable Excel alerts

    for filename in os.listdir(directory):
        if filename.endswith(".xls") or filename.endswith(".XLS"):
            file_path = os.path.join(directory, filename)
            new_file_path = os.path.splitext(file_path)[0] + ".xlsm"

            # Open the file and check if a workbook object was successfully created
            workbook = excel.Workbooks.Open(file_path)
            if workbook is not None:
                workbook.SaveAs(new_file_path, FileFormat=52)  # Save as .xlsm
                workbook.Close(SaveChanges=False)  # Close the workbook without saving changes
                print(f"Converted {filename} to {os.path.basename(new_file_path)}")
            else:
                print(f"Failed to open {filename}")

    excel.Quit()  # Quit Excel

# Example usage:
convert_xls_to_xlsm(directory_path)
```
